## Download youtube videos and store as .webm

```python
import subprocess
import os

# Create the 'songs' directory if it doesn't exist
os.makedirs('../canto_songs', exist_ok=True)

# Read the YouTube links, song names, and albums from the text file
with open('../sample_data/youtube_links.txt', 'r', encoding='utf-8') as file:  # Ensure UTF-8 encoding for mixed languages
    lines = file.readlines()

# Loop through each line, split the link, artist, song name, and album, and download the corresponding video
for line in lines:
    line = line.strip()  # Remove any leading/trailing whitespace
    if line:
        # Split the line into YouTube link, artist, song name, and album
        parts = line.split(',', 3)  # Split into 4 parts
        if len(parts) == 4:
            link, artist, song_name, album = [part.strip() for part in parts]  # Clean up extra spaces around each part

            # Construct the output file path with album and song name
            output_path = f'../canto_songs/{artist}_{album}_{song_name}.%(ext)s'

            # Download the video using yt-dlp and save it as the song name in mp3 format
            subprocess.run([
                'yt-dlp', link,
                '--format', 'bestaudio',
                '--extract-audio',
                '--audio-format', 'mp3',
                '--no-check-certificate',  # Bypass SSL verification
                '-o', output_path
            ])
```

## Convert .webm to .mp3 (if required)

```python
import subprocess
import os

# Define the input directory containing the .webm files
input_directory = '../canto_songs'

# Define the output directory for the converted .mp3 files
output_directory = '../converted_canto_songs'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

try:
    # Get a list of all .webm files in the input directory
    webm_files = [file_name for file_name in os.listdir(input_directory) if file_name.endswith('.webm')]

    # Check if there are any .webm files
    if not webm_files:
        print("No .webm files found in the directory.")
    else:
        # Process each .webm file
        for file_name in webm_files:
            print(f"Processing file: {file_name}")

            # Construct the full input file path
            input_path = os.path.join(input_directory, file_name)

            # Construct the output file path in the output directory
            output_path = os.path.join(output_directory, file_name.replace('.webm', '.mp3'))

            try:
                # Use ffmpeg to convert the .webm file to .mp3
                subprocess.run([
                    'ffmpeg', '-i', input_path,  # Input file
                    '-vn',                       # No video
                    '-ar', '44100',              # Set audio sample rate
                    '-ac', '2',                  # Set number of audio channels
                    '-b:a', '192k',              # Set audio bitrate
                    output_path                  # Output file
                ], check=True)  # `check=True` raises an exception if ffmpeg fails

                print(f"Successfully converted: {file_name} -> {os.path.basename(output_path)}")
            except subprocess.CalledProcessError as e:
                print(f"Error processing file {file_name}: {e}")

except FileNotFoundError as e:
    print(f"Error: Input directory '{input_directory}' not found. Please check the directory path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
