## Transcription using base whisper-small model

```python
import os

# Set the directory containing the songs
songs_dir = '../canto_songs'

# Set the output directory for transcriptions
output_dir = '../transcribed_canto_songs'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# List all .mp3 files in the directory
songs = [f for f in os.listdir(songs_dir) if f.endswith('.mp3')]

# Loop through each song and transcribe
for song in songs:
    song_path = os.path.join(songs_dir, song)
    output_file = os.path.join(output_dir, os.path.splitext(song)[0] + '.srt')  # Save as .srt file

    print(f"Transcribing {song}...")

    # Use the `os.system` command to call Whisper with the appropriate output file
    os.system(f'whisper "{song_path}" --model small --language Chinese --output_dir "{output_dir}"')

    # Check if the transcription was created
    if os.path.exists(output_file):
        print(f"Transcription saved with timestamps to: {output_file}")
    else:
        print(f"Failed to transcribe: {song}")

print("All songs have been transcribed.")
```

```python
from transformers import pipeline

# Load the Whisper model pipeline
pipe = pipeline(model="Maverickz1989/openai-whisper-large-v2-colab")

# Transcription function
def transcribe(audio):
    text = pipe(audio)["text"]
    return text
```

```python
import os
from transformers import pipeline

# Load the Whisper model pipeline
pipe = pipeline(model="Maverickz1989/openai-whisper-large-v2-colab")

# Function to split long audio transcription into SRT format
def transcribe_and_save_as_srt(audio_path, output_path):
    print(f"Transcribing: {audio_path}")

    # Perform transcription with timestamps
    try:
        result = pipe(audio_path, return_timestamps=True)  # Enable timestamps
        transcription_segments = result["chunks"]  # Extract transcription chunks with timestamps

        # Create SRT content
        srt_content = ""
        for idx, segment in enumerate(transcription_segments, start=1):
            start_time = format_timestamp(segment["timestamp"][0])  # Start timestamp
            end_time = format_timestamp(segment["timestamp"][1])  # End timestamp
            text = segment["text"]

            # Format each segment as an SRT entry
            srt_content += f"{idx}\n{start_time} --> {end_time}\n{text.strip()}\n\n"

        # Write the transcription to the output SRT file
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(srt_content)

        print(f"Transcription saved to: {output_path}")
    except Exception as e:
        print(f"Error transcribing {audio_path}: {e}")

# Helper function to format timestamps for SRT
def format_timestamp(seconds):
    milliseconds = int((seconds % 1) * 1000)
    seconds = int(seconds)
    minutes, seconds = divmod(seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return f"{hours:02}:{minutes:02}:{seconds:02},{milliseconds:03}"

# Directory containing the songs
songs_dir = './canto_songs'

# Directory to save the transcriptions
output_dir = './transcribed_canto_songs_finetune'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# List all .mp3 files in the input directory
songs = [f for f in os.listdir(songs_dir) if f.endswith('.mp3')]

# Loop through each song, transcribe, and save as SRT
for song in songs:
    song_path = os.path.join(songs_dir, song)
    output_file = os.path.join(output_dir, os.path.splitext(song)[0] + '.srt')  # Save as .srt file

    print(f"Processing {song}...")
    transcribe_and_save_as_srt(song_path, output_file)

print("All songs have been transcribed.")
```
